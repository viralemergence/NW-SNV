```python
from Bio import SeqIO
import Bio.SeqUtils as SeqUtils
import pandas as pd

#No need to find reverse complement since linear RNA and cDNA have the same sequence, with T instead of U.


def get_gc_content(seq):
    return SeqUtils.gc_fraction(seq)

fasta_path = '../Refseq/SNV_Segments_refseq.fasta'
records = SeqIO.parse(fasta_path, "fasta")

for record in records:
    seq = record.seq
    gc_content = get_gc_content(seq)
    print(f'{record.id} \t %GC = {round(gc_content,3)}, %AT = {round(1-gc_content,3)}')



```
